```python
#Librerias
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
#Carga
df_1 = pd.read_csv("DATA/puntajes2023.csv", sep=';', low_memory=False, decimal=",")
df_2 = pd.read_csv("DATA/puntajes2024.csv", sep=';', low_memory=False, decimal=",")
df_3 = pd.read_csv("DATA/puntajes2025.csv", sep=';', low_memory=False, decimal=",")
```

```python
idx = df_2.columns.get_loc("PRUEBA_REG_CL")
df_2 = df_2.iloc[:, :idx]

delet = df_2.loc[:,'CLEC_REG_ACTUAL': 'CIEN_INV_ANTERIOR'].columns
df_2 = df_2.drop(columns=delet)
df_2
```

```python
#Limpieza #1
idx = df_3.columns.get_loc("PRUEBA_REG_CL")
df_3 = df_3.iloc[:, :idx]

delet = df_3.loc[:,'CLEC_REG_ACTUAL': 'CIEN_INV_ANTERIOR'].columns
df_3 = df_3.drop(columns=delet)
df_3
```

```python
df_1
```

```python
#Limpieza de valores nulos
df_1.replace(['', ' '], np.nan, inplace=True)
df_2.replace(['', ' '], np.nan, inplace=True)
df_3.replace(['', ' '], np.nan, inplace=True)
df_1.dropna(inplace=True)
df_2.dropna(inplace=True)
df_3.dropna(inplace=True)
```

```python
df = pd.concat([df_1, df_2, df_3], axis=0, ignore_index=True)
df
```

```python
df = df[df['PROMEDIO_CM_MAX'] != 0]
```

```python
#Agregations
df['DEPENDENCIA'] = df['DEPENDENCIA'].astype(int)
reclasificación = {
    1: 1, 2: 1, 6: 1,  #Gratuitos
    3: 2, 5: 2,        #Subvencionados
    4: 3               #Particular 
}
df["FINANCIAMIENTO"] = df['DEPENDENCIA'].map(reclasificación)
df['FINANCIAMIENTO']=df['FINANCIAMIENTO'].astype(object)
df.groupby('FINANCIAMIENTO').agg({'PROMEDIO_CM_MAX': 'mean'})
```

```python
df.describe()
```

```python
#Grafico 1

sns.set_style('white')
plt.figure(figsize=(13, 10))
g1 = sns.histplot(x='PROMEDIO_CM_MAX',
                  bins= 20,
                  color='blue',
                  edgecolor= 'black',
                  alpha=0.3,
                  data=df)

g1.set_title('Distribución de puntajes promedio PAES', fontweight = 'bold',fontsize = 16, pad=20)
g1.set_xlabel('Puntaje promedio', fontweight='bold', fontsize = 14)
g1.set_ylabel('Frecuencia',fontweight='bold', fontsize = 14)

mean_value = df['PROMEDIO_CM_MAX'].mean()
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Media: {mean_value:.2f}')
plt.legend()

# Rango y ticks del eje X
min_x = 100
max_x = int(df['PROMEDIO_CM_MAX'].max()) + 10
plt.xlim(min_x, max_x)
plt.xticks(np.arange(min_x, max_x + 1, 50))  # cada 50
plt.xticks(fontsize=13)
plt.yticks(fontsize=13) 

plt.xlim(100, df['PROMEDIO_CM_MAX'].max() + 10)

sns.despine()
plt.savefig('grafico_1.png', dpi=600)
```

```python
#Grafico 2
sns.set_style('white')
plt.figure(figsize=(13, 10))
g2 = sns.boxplot(x= 'FINANCIAMIENTO',
                 y= 'PROMEDIO_CM_MAX',
                 data=df,
                 palette='viridis',
                 boxprops=dict(alpha=0.92),
                 )

g2.set_title('Puntaje segun tipo de financiamiento de la institución', fontweight='bold', fontsize = 16, pad= 20)
g2.set_xticklabels(['Gratuito', 'Subvencionado', 'Particular'])
g2.set_xlabel('Tipo de Financiamiento de institución', fontweight='bold', fontsize = 14)
g2.set_ylabel('Puntaje promedio', fontweight='bold', fontsize = 14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
sns.despine()
plt.savefig('grafico_2.png', dpi=600)
plt.show()
```

```python
#Grafico 3
plt.figure(figsize=(13, 10))
sns.set_style("white")

g3 = sns.lineplot(
    x='ANYO_PROCESO',
    y='PROMEDIO_CM_MAX',
    hue='FINANCIAMIENTO',
    data=df,
    markers='True',
    linewidth=4,
    palette='viridis',
    errorbar=None,
)
plt.xticks(sorted(df['ANYO_PROCESO'].unique()))

for line in range(len(g3.lines)):
    x_data = g3.lines[line].get_xdata()
    y_data = g3.lines[line].get_ydata()
    
    for i in range(len(x_data)):
        x, y = x_data[i], y_data[i]
        
        if i == 0:
            continue
        else:
            prev_y = y_data[i-1]
            pct_change = ((y - prev_y) / prev_y) * 100
            if pct_change > 0:
                pct_text = f'+{pct_change:.1f}%'
                color = 'green'
            else:
                pct_text = f'{pct_change:.1f}%'
                color = 'red'
                
            g3.annotate(pct_text, 
                       (x, y), 
                       textcoords="offset points",
                       xytext=(0, 10), 
                       ha='center', 
                       fontsize=11,
                       fontweight='bold',
                       color=color)
            
g3.set_xlabel('Año', fontweight='bold', fontsize = 14)
g3.set_ylabel('Puntaje promedio', fontweight='bold', fontsize = 14)
g3.set_title('Puntajes segun financiamiento a través del tiempo',fontweight='bold', fontsize = 16, pad= 20)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(title='Tipo de financiamiento', labels = ['Gratuito', 'Subvencionado', 'Privado'], loc ='lower right')
plt.ylim(500, 770)
sns.despine()
plt.savefig('grafico_3.png', dpi=600)
```
